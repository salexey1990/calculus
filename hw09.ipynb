# 2

```python
def equations(p):
    x, y, z = p
    return (
        4*x+12*y+2*x*z,12*x+64*y+32*y*z,x**2+16*y**2-64
    )

# Численное решение нелинечной системы уравнений
x, y, z =  fsolve(equations, (1, 1, 1))
print (x, y, z)
```

```python
def equations_2(p):
    x, y = p
    return (
        x**2-y**2+3*x*y**3-2*x**2*y**2+2*x-3*y-5,
        3*y**3-2*x**2+2*x**3*y-5*x**2*y**2+5
    )

# Численное решение нелинечной системы уравнений
for i in np.arange(-20, 20, 1):
    
    x, y =  fsolve(equations_2, (i, i))
    print (x, y, i)
```
